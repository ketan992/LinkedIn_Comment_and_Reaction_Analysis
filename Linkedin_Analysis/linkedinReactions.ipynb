```python
from bs4 import BeautifulSoup
import pandas as pd

with open("reactions.html", encoding="utf-8") as fp:
    soup = BeautifulSoup(fp, "html.parser")


# Find all the li tags with the desired class
li_tags = soup.find_all("li", class_="social-details-reactors-tab-body-list-item")

# Create an empty DataFrame with the desired column names
df = pd.DataFrame(columns=["Name", "Job Title", "Reaction","Profile Url"])

# Loop over the li tags and extract the desired information from each one
for li_tag in li_tags:
    name = li_tag.find("span", dir="ltr").text.strip()
    job_title = li_tag.find("div", class_="artdeco-entity-lockup__caption").text.strip()
    like_icon = li_tag.find("img", class_="reactions-icon")
    like_icon_alt = like_icon.get("alt", "")
    link  = li_tag.find("a",class_="link-without-hover-state ember-view")
    link_value = link.get("href", "")
    # Append the extracted information to the DataFrame
    df = df.append({"Name": name, "Job Title": job_title, "Reaction": like_icon_alt,"Profile Url":link_value}, ignore_index=True)
df
```

```python
df.to_csv('LinkedinReactionsv3.csv',index = False)
```

```python
one_hot = pd.get_dummies(df['Reaction'])

# Add the one hot encoded columns to the original dataframe
df = pd.concat([df, one_hot], axis=1)

df
```
