```python
df1['Department'] = ''
for i in range(len(df1['jobTitle'])) :
    department = 'Others'
    
    for key, value in dict1.items() :
        
        if df1['jobTitle'][i] in value : 
            department = key
            break
    df1['Department'][i] = department

        
```

```python
import pandas as pd
import matplotlib.pyplot as plt

funny = df1[df1['Classification'] == 'Negative']

counts = funny['Department'].value_counts()

plt.figure(figsize=(10, 5), dpi=150)
plt.bar(counts.index, counts.values)
plt.title('Department wise count for Negative Comment Sentiment on Linkedin post')
plt.xlabel('Department')
plt.ylabel('Count')
plt.xticks(rotation=90)
plt.show()
```
