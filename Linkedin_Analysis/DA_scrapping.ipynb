```python
src = driver.page_source
soup = BeautifulSoup(src, 'lxml') #comments-comment-item comments-comments-list__comment-item
```

```python
comments = soup.findAll('article', {'class': 'comments-comments-list__comment-item'})
```

```python
comments
```

```python
main_comment = []
replies = []
count = 0
for d in comments:
    reply = d.find('div', {'class': 'comments-comment-item__replies-list'})
    replies.append(reply)
```

```python
replies[0]
```

```python
len(replies)
```

```python
all_replies = []
each_reply = replies[0].findAll('article', {'class': 'comments-comment-item'})
all_replies.append(each_reply)
```

```python
len(all_replies[0])
```

```python
all_replies[0]
```

```python
profiles = []
for each_comm in all_replies[0]:
    user_comment_map = {}
    comm_user = each_comm.find('a', {'class': 'comments-post-meta__actor-link'}, href=True)
    comm_made = each_comm.find('div', {'class': 'update-components-text'})
    if comm_user['href'] in user_comment_map:
        print("Append made here", comm_user['href'])
        user_comment_map[comm_user['href']].append(comm_made.text.strip())
    user_comment_map[comm_user['href']] = [comm_made.text.strip()]
#     user_comment_map['profile_url'] = comm_user['href']
#     user_comment_map['comment'] = comm_made.text.strip()
    profiles.append(user_comment_map)
```

```python
updated_profiles = []
for obj in profiles:
    for key in obj:
        prof = {}
        prof['profile_url'] = key
        prof['comment'] = str(obj[key]).encode('UTF-8')
        updated_profiles.append(prof)
```

```python
updated_profiles
```

```python
myFile = open('profile_to_comment.csv', 'w')
writer = csv.writer(myFile)
writer.writerow(['profile_url', 'comment'])
for obj in updated_profiles:
    writer.writerow(obj.values())
myFile.close()
```
