```python
departments_keywords = {
    'Management & Senior-Level': ['Manager', 'Director', 'CEO', 'President', 'Founder', 'Head', 'Supervisor', 'Lead', 'Chief', 'Senior', 'Sr.'],
    'Sales': ['Sales', 'Business Development', 'KAM'],
    'Human Resources': ['Human Resource', 'HR', 'Recruitment', 'Compensation'],
    'Engineering': ['Engineer', 'Developer', 'DevOps'],
    'IT': ['Information Technology', 'IT', 'CTO', 'Support', 'Technician'],
    'Operations': ['Operations', 'Logistics', 'Procurement', 'Sourcing'],
    'Education': ['Teaching', 'Learning', 'Mentor', 'Fellow'],
    'Other': []
}
```

```python
def categorize_job_titles_single_department(unique_jobtitle, departments_keywords):
    categorized_jobs = {department: [] for department in departments_keywords}

    for title in unique_jobtitle:
        department_found = False
        for department, keywords in departments_keywords.items():
            if any(keyword in title for keyword in keywords):
                categorized_jobs[department].append(title)
                department_found = True
                break  # Stop searching for other matches once a department is found
        if not department_found:
            categorized_jobs['Other'].append(title)

    return categorized_jobs

categorized_jobs = categorize_job_titles_single_department(unique_jobtitle, departments_keywords)

for department, titles in categorized_jobs.items():
    print(f"{department}:")
    print(titles)
    print("\n")
```

```python
def map_title_to_department(job_title, categorized_jobs):
    for department, titles in categorized_jobs.items():
        if job_title in titles:
            return department
    return 'Other'

# Replace the values in the 'jobTitle' column with the department keys
df1['Department'] = df1['jobTitle'].apply(lambda title: map_title_to_department(title, categorized_jobs))
df1['Department']
```

```python
import pandas as pd
import statsmodels.api as sm
from statsmodels.graphics.mosaicplot import mosaic
import matplotlib.pyplot as plt
```

```python
def create_mosaic_plot(df, col1, col2,cmap='viridis'):
    count_df = df.groupby([col1, col2]).size().reset_index(name='count')
    total_count = count_df['count'].sum()
    count_df['proportion'] = count_df['count'] / total_count
    pivot_df = count_df.pivot_table(index=col1, columns=col2, values='proportion', fill_value=0)
    plt.figure(figsize=(12, 8))
    mosaic_plot = mosaic(pivot_df.stack(), title=f'Mosaic Plot for {col1} and {col2}')
    plt.show()
```

```python
unique_countries = df1['country'].unique()

for country in unique_countries:
    print(f"Mosaic Plot for {country}")
    filtered_df = df1[df1['country'] == country]
    create_mosaic_plot(filtered_df, 'country', 'jobTitle')
```
